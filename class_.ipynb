```python
### 문제 1
이름(`name`)과 학번(`student_id`)을 속성으로 가지는 `Student` 클래스를 작성하세요.
1. `__init__` 생성자를 사용하여 두 속성을 초기화하세요.
2. 본인의 이름과 학번을 가진 인스턴스를 생성하여 각 속성값을 출력하세요.
```

```python
# 여기에 코드를 작성하세요.

class Student:
    def __init__(self,student_name,student_number):
        self.name = student_name
        self.number = student_number
student1 = Student("파기", 2026)       
print(student1.name)
print(student1.number)
```
